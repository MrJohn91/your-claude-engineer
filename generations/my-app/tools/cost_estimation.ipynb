## Instagram Cost Estimation

```python
# Estimate cost for Instagram profiles
num_instagram_profiles = 100
instagram_cost = estimate_instagram_cost(num_instagram_profiles)
print(f"Cost for {num_instagram_profiles} Instagram profiles:")
print(instagram_cost)
```

## Total Cost Estimation

```python
# Calculate total cost
total_cost = estimate_total_cost(num_linkedin_profiles, num_instagram_profiles)
print(f"Total cost for {num_linkedin_profiles} LinkedIn + {num_instagram_profiles} Instagram profiles:")
print(total_cost)
```
